Loads the ontology and displays all the Persons (without using queries, without inference).

```python
from rdflib import Graph
```

```python
# Create a Graph
g = Graph()
```

```python
g.parse("withpeeps.rdf")
```

```python
# Loop through each triple in the graph (subj, pred, obj)
for subj, pred, obj in g:
    # Check if there is at least one triple in the Graph
    if (subj, pred, obj) not in g:
       raise Exception("It better be!")

# Print the number of "triples" in the Graph
print(f"Graph g has {len(g)} statements.")

# Print out the entire Graph in the RDF Turtle format
print(g.serialize(format="n3"))
```

```python
webSem = Namespace("http://www.semanticweb.org/marya/ontologies/2023/2/webSem#")
```

```python
for s, p, o in g.triples((None, RDF.type, webSem.Person)):
    print(f"{s} is a person")
```

Loads the ontology and displays all the Persons (using a query, without inference). Create the used query in text file under the data folder.

```python
qres = g.query("""
                SELECT ?name
                    WHERE { ?x rdf:type :Person.
                             ?x  :name ?name }
                """)

print(f"Number of persons : {len(qres)}")
for row in qres:
    print(f"{row.name}")
```

Develops a program that :
- Reads a name of a movie
- If it doesn’t exist displays an error message
- Else, display its year, country, genres and actors
- Display their program where and when

```python
def searchMovie(name):
    qres = g.query("""
                SELECT ?title ?year ?country ?genre ?actor ?start ?end
                    WHERE { ?x :title """ + "'"+name +"'" + """.
                            ?x :title ?title.
                            ?x :year ?year.
                            ?x :country ?country.
                            ?x :genre ?genre.
                            OPTIONAL{
                                ?x :isPlayedBy :Person.
                                ?y :name ?actor.
                                ?x :inTheDatesOf ?a.
                                ?z :hasBroadcastTimeOf ?a.
                                ?a :startDate ?start.
                                ?a :endDate ?end
                            }
                            }
                """)
    
    genre = ""
    if len(qres) != 0:
        for row in qres:
            genre += row.genre + " "
            rep = f"title : {row.title}, year : {row.year}, country : {row.country}, actors : {row.actor}, {row.start}, {row.end}"

        print(rep, "genres : ",genre)
    else:
        print("Error! No movies found.")
```

```python
x = input()
```

```python
searchMovie(x)
```

Displays all persons that are actors and directors. Do this using a rule that defines a new class ActorDirector. The rule file must be saved in the data folder.

```python
with onto:
    rule = Imp()
    rule.set_as_rule("""wasDirectedBy(?m, ?x), isPlayedBy(?m, ?x) -> ActorDirector(?x)""")
```

```python
for actor in onto.search(type = onto.ActorDirector):
    print(actor)
```

Specifies 3 different rules and implement them in a java program

```python
with onto:
    rule1 = Imp()
    rule1.set_as_rule("""isActorOf(?x, ?m), isActorOf(?y, ?m) -> hasActedWith(?x, ?y)""")
    
    rule2 = Imp()
    rule2.set_as_rule(""" Movie(?x) ^ year(?x, ?y) ^ lessThan(?y, 1995) -> OldMovie(?x)""")
    
    rule3 = Imp()
    rule3.set_as_rule(""" Movie(?x) ^ country(?x, ?r) ^ contains(?r, "France") -> FrenchMovie(?x)""")
```

```python
for actor in onto.search(hasActedWith = onto.BradPitt):
    print(actor)
```

```python
for movie in onto.search(type = onto.OldMovie):
    print(movie)
```

```python
for movie in onto.search(type = onto.FrenchMovie):
    print(movie)
```
