```python
#Defining the namespace
nspace = Namespace("Project.rdf")

# Create a Graph
g = Graph()

# Parse in an RDF file hosted locally
g.parse("projet3.ttl")

g.namespace_manager.bind("nl",nspace)

# Loop through each triple in the graph (subj, pred, obj)
for subj, pred, obj in g:
    # Check if there is at least one triple in the Graph
    if (subj, pred, obj) not in g:
       raise Exception("It better be!")

# Print the number of "triples" in the Graph
print(f"Graph g has {len(g)} statements.")

# Print out the entire Graph in the RDF Turtle format
print(g.serialize(format="n3"))
```

```python
for s, p, o in g:
    print(s,p,o)
```

```python
properties = set()

for s, p, o in g:
    properties.add(s)

for p in properties:
    print(p)
```

```python
for prefix, ns in g.namespaces():
    print(prefix, ns)
```

# Insert Movies

```python
for movie in data:
    for genre in movie["Genre"]:
        print(genre)
```

```python
for movie in data:
    #print(movie)
    ref = websem[movie["Title"].replace(" ","")]
    #print(ref)
    title = Literal(movie["Title"])
    year = Literal(movie["Year"], datatype=XSD["int"])
    country = Literal(movie["Country"])
    language = Literal(movie["Language"])
    overview = Literal(movie["Plot"])
    rated = Literal(movie["Rated"])
    imdbRating = Literal(movie["imdbRating"], datatype=XSD["float"])
    
    g.add((ref, RDF.type, websem["Movie"]))
    g.add((ref, websem["title"], title))
    g.add((ref, websem["year"], year))
    g.add((ref, websem["country"], country))
    g.add((ref, websem["language"], language))
    g.add((ref, websem["overview"], overview))
    g.add((ref, websem["rated"], rated))
    g.add((ref, websem["imdbRating"], imdbRating))
    
    for i in movie["Genre"]:
        genre = Literal(i)
        g.add((ref, websem["genre"], genre))
    
    # add the Object properties. 
    for d in movie["Writer"]:
        g.add((ref, websem["wasDirectedBy"], websem[d.replace(" ","")]))
    #g.add((ref, websem["wasDirectedBy"], websem[movie["Director"].replace(" ","")]))
    
    for w in movie["Writer"]:
        g.add((ref, websem["wasWrittenBy"], websem[w.replace(" ","")]))
    
    for a in movie["Actors"]:
        g.add((ref, websem["isPlayedBy"], websem[a.replace(" ","")]))
```

```python
print(g.serialize(format="n3"))
```

```python
g.serialize(destination="withpeeps.rdf")
```

# Queries

List the instances of the class Actor

```python
qres = g.query("""
                SELECT ?name
                    WHERE { ?x rdf:type webSem:Actor.
                             ?x  webSem:name ?name }
                """)


for row in qres:
    print(f"{row.name}")
```

List the name of all Thriller movies. For each one, display its director.

```python
qres = g.query("""
                SELECT ?title ?director
                    WHERE { ?x webSem:wasDirectedBy ?y.
                            ?y webSem:name ?director.
                            ?x webSem:title ?title.
                            ?x  webSem:genre "Thriller" }
                """)


for row in qres:
    print(f"{row.title}, {row.director}")
```

List the name of all Crime Thriller movies.

```python
qres = g.query("""
                SELECT ?title ?genre
                    WHERE { ?x webSem:title ?title.
                            ?x  webSem:genre "Thriller".
                            ?x webSem:genre "Crime".
                            ?x webSem:genre ?genre
                            }
                """)


for row in qres:
    print(f"{row.title}, {row.genre}")
```

List the name of Actors older than 51 years.

```python
qres = g.query("""
                SELECT ?name ?age
                    WHERE { ?x rdf:type webSem:Actor.
                             ?x  webSem:name ?name.
                             ?x webSem:age ?age 
                             FILTER (?age>51)}
                """)


for row in qres:
    print(f"{row.name}, {row.age}")
```

List of movies that are played in theater for a specific day and where and until when

```python
qres = g.query("""
                SELECT ?title ?start ?end
                    WHERE { ?x rdf:type webSem:Movie.
                            ?x webSem:title ?title.
                             ?x  webSem:inTheDatesOf ?y.
                             ?z webSem:hasBroadcastTimeOf ?y.
                             ?y webSem:startDate ?start.
                             ?y webSem:endDate ?end
                             }

                """)


for row in qres:
    print(f"{row.title}, {row.start}, {row.end}")
```

A query that contains at least 2 Optional Graph Patterns

```python
qres = g.query("""
                SELECT ?name ?age ?nationality
                    WHERE { ?x rdf:type webSem:Person.
                             ?x  webSem:name ?name.
                             OPTIONAL{?x webSem:age ?age.
                                        ?x webSem:nationalty ?nationality}
                             }
                """)


for row in qres:
    print(f"{row.name}, {row.age}, {row.nationality}")
```

A query that contains at least 2 alternatives and conjunctions

A query that contains a CONSTRUCT query form

```python
qres = g.query("""
                CONSTRUCT { ?x webSem:name ?name }
                WHERE {?x webSem:name ?name }
                """)


for row in qres:
    print(f"{row[0]}, {row[1]}, {row[2]}")
```

A query that contains an ASK query form

```python
qres = g.query("""
                ASK { ?x webSem:name "Brad Pitt" }
                """)


for row in qres:
    print(row)
```

A query that contains a DESCRIBE query form

```python
qres = g.query("""
                DESCRIBE ?x WHERE {
                ?x webSem:name "Brad Pitt"
                }
                """)


for row in qres:
    print(row)
```
